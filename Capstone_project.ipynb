```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (single file with both red and white wines)
wine_data = pd.read_csv('winequalityN.csv')

# Check if 'type' column exists; if not, we need to infer it
# If you know that the first few rows correspond to red wine and others to white wine, for example:
wine_data['type'] = wine_data['type'].str.strip()  # Cleaning the 'type' column if necessary

# Check for unique values in the 'type' column to verify
print(wine_data['type'].unique())

# Set up the figure for the histogram
plt.figure(figsize=(10, 6))

# Plot the histogram for red wines
wine_data[wine_data['type'] == 'red']['quality'].hist(bins=range(3, 11), alpha=0.5, label='Red Wine', color='red', edgecolor='black')

# Plot the histogram for white wines
wine_data[wine_data['type'] == 'white']['quality'].hist(bins=range(3, 11), alpha=0.5, label='White Wine', color='yellow', edgecolor='black')

# Add titles and labels
plt.title('Wine Quality Distribution by Type')
plt.xlabel('Quality')
plt.ylabel('Frequency')
plt.legend()

# Show the plot
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (make sure to adjust the path if necessary)
wine_data = pd.read_csv('winequalityN.csv')

# Check the first few rows to understand the structure of the data
wine_data.head()

# Select only numeric columns to compute correlations (exclude non-numeric columns like 'type')
numeric_columns = wine_data.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix for the numeric columns
correlation_matrix = numeric_columns.corr()

# Print the correlation between features and the target variable 'quality'
correlation_with_quality = correlation_matrix['quality'].sort_values(ascending=False)

# Display the top 5 features most correlated with quality (excluding 'quality' itself)
print("Top features contributing to quality:")
print(correlation_with_quality[1:6])

# Optional: Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
import pandas as pd

wine_data = pd.read_csv('winequalityN.csv')

def search_filter(wine_data, type=None, min_fixed_acidity=None, max_fixed_acidity=None, 
                  min_volatile_acidity=None, max_volatile_acidity=None, 
                  min_citric_acid=None, max_citric_acid=None, 
                  min_residual_sugar=None, max_residual_sugar=None, 
                  min_chlorides=None, max_chlorides=None,
                  min_sulfur_dioxide=None, max_sulfur_dioxide=None, 
                  min_density=None, max_density=None, 
                  min_pH=None, max_pH=None, 
                  min_sulphates=None, max_sulphates=None, 
                  min_alcohol=None, max_alcohol=None, 
                  min_quality=None, max_quality=None):
    
    filtered_data = wine_data
    

    if type:
        filtered_data = filtered_data[filtered_data['type'] == type]
    

    if min_fixed_acidity is not None:
        filtered_data = filtered_data[filtered_data['fixed acidity'] >= min_fixed_acidity]
    if max_fixed_acidity is not None:
        filtered_data = filtered_data[filtered_data['fixed acidity'] <= max_fixed_acidity]
    

    if min_volatile_acidity is not None:
        filtered_data = filtered_data[filtered_data['volatile acidity'] >= min_volatile_acidity]
    if max_volatile_acidity is not None:
        filtered_data = filtered_data[filtered_data['volatile acidity'] <= max_volatile_acidity]
    

    if min_citric_acid is not None:
        filtered_data = filtered_data[filtered_data['citric acid'] >= min_citric_acid]
    if max_citric_acid is not None:
        filtered_data = filtered_data[filtered_data['citric acid'] <= max_citric_acid]
    

    if min_residual_sugar is not None:
        filtered_data = filtered_data[filtered_data['residual sugar'] >= min_residual_sugar]
    if max_residual_sugar is not None:
        filtered_data = filtered_data[filtered_data['residual sugar'] <= max_residual_sugar]
    

    if min_chlorides is not None:
        filtered_data = filtered_data[filtered_data['chlorides'] >= min_chlorides]
    if max_chlorides is not None:
        filtered_data = filtered_data[filtered_data['chlorides'] <= max_chlorides]
    

    if min_sulfur_dioxide is not None:
        filtered_data = filtered_data[filtered_data['free sulfur dioxide'] >= min_sulfur_dioxide]
    if max_sulfur_dioxide is not None:
        filtered_data = filtered_data[filtered_data['free sulfur dioxide'] <= max_sulfur_dioxide]
    

    if min_sulfur_dioxide is not None:
        filtered_data = filtered_data[filtered_data['total sulfur dioxide'] >= min_sulfur_dioxide]
    if max_sulfur_dioxide is not None:
        filtered_data = filtered_data[filtered_data['total sulfur dioxide'] <= max_sulfur_dioxide]
    

    if min_density is not None:
        filtered_data = filtered_data[filtered_data['density'] >= min_density]
    if max_density is not None:
        filtered_data = filtered_data[filtered_data['density'] <= max_density]
    

    if min_pH is not None:
        filtered_data = filtered_data[filtered_data['pH'] >= min_pH]
    if max_pH is not None:
        filtered_data = filtered_data[filtered_data['pH'] <= max_pH]
    
    if min_sulphates is not None:
        filtered_data = filtered_data[filtered_data['sulphates'] >= min_sulphates]
    if max_sulphates is not None:
        filtered_data = filtered_data[filtered_data['sulphates'] <= max_sulphates]
    

    if min_alcohol is not None:
        filtered_data = filtered_data[filtered_data['alcohol'] >= min_alcohol]
    if max_alcohol is not None:
        filtered_data = filtered_data[filtered_data['alcohol'] <= max_alcohol]
    
    if min_quality is not None:
        filtered_data = filtered_data[filtered_data['quality'] >= min_quality]
    if max_quality is not None:
        filtered_data = filtered_data[filtered_data['quality'] <= max_quality]
    
    return filtered_data


def take_user_input():
    print("Enter filter criteria for wine search.")
    
    wine_type = input("Wine Type (red/white): ").strip().lower()
    if wine_type not in ['red', 'white']:
        print('Wrong Input')
        return
    wine_type = wine_type if wine_type in ['red', 'white'] else None  
    
    min_quality = input("Minimum Quality (0-10): ")
    max_quality = input("Maximum Quality (0-10): ")
    min_quality = int(min_quality) if min_quality else None
    max_quality = int(max_quality) if max_quality else None

    min_alcohol = input("Minimum Alcohol (0-20): ")
    max_alcohol = input("Maximum Alcohol (0-20): ")
    min_alcohol = float(min_alcohol) if min_alcohol else None
    max_alcohol = float(max_alcohol) if max_alcohol else None

    results = search_filter(
        wine_data,
        type=wine_type,
        min_quality=min_quality,
        max_quality=max_quality,
        min_alcohol=min_alcohol,
        max_alcohol=max_alcohol
    )

    return results

filtered_results = take_user_input()
print("Filtered results:")
print(filtered_results)
```
